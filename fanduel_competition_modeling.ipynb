## Preparing Data for XGBoost

````python
#Player DF Dict

# display(dfx.info(),dfx.head())

## Preparing Data for XGBoost
import re
def sep_teams(row):

    import re
    row['teams'] = row['teams'].strip()
    matches = re.findall(r"([A-Z]+) .+ ([A-Z]+)", row['teams'])[0]
#     display(matches)

    row['team_1'] = matches[0]#[0]#re.findall(r'(\w*).[@|(vs)].(\w*)',row['teams'])[0][0]
    row['team_2'] = matches[-1]#[1]#re.findall(r'(\w*).[@|(vs)].(\w*)',row['teams'])[0][1]
    return row
def prep_player_nba_stats_df(PLAYER_ROW,target_col = 'total_pts'):
    ## Target column for regression is total points
    dfx = pd.read_csv(PLAYER_ROW['player_csv'])
    dfx = ns.calc_fantasy_stats(dfx)
#     dfx = dfx.set_index('date')
    
    ## Get list of fantasy points columns to keep
    fpoints_cols = [col for col in dfx.columns if 'pts' in col]

    ## Get list of original cols for fantasy points to drop
    orig_cols = [col.replace('_pts','') for col in fpoints_cols]
    orig_cols.remove('total')


    ## columns to drop for remaining column names
    drop_cols = orig_cols
    drop_cols.extend(fpoints_cols)

    ## Columns to keep
    keep_cols = fpoints_cols
    keep_cols.extend(dfx.drop(drop_cols,axis=1).columns)
    # orig_cols,fpoints_cols
    # keep_cols

    dfx = dfx[keep_cols].set_index('player_id')
    
    ## Separate Teams inton 2 features
    dfx = dfx.apply(sep_teams,axis=1)
    
    dfx.set_index('date',inplace=True)


    return dfx
dfx = prep_player_nba_stats_df(PLAYER,)    
dfx.head(3)
````

````python
def train_test_index(X,y,n_test_games=2):
    
    if isinstance(X,pd.Series):
        X = X.values
        
    if isinstance(y,pd.Series):
        y=y.values
    X_train = X[:-n_test_games]#.copy()
    X_test = X[-n_test_games:]#.copy()
    
    y_train = y[:-n_test_games]#.copy()
    y_test = y[-n_test_games:]#.copy()
    
        

    
    return X_train,X_test,y_train,y_test
X_train,X_test,y_train,y_test = train_test_index(X,y,n_test_games=2)
X_train.shape, X_test.shape,y_train.shape,y_test.shape
````
